```python
from tkinter import *
```

```python
win = Tk()
win.title("Daum Log-In")
win.geometry("350x250")
win.option_add("*Font", "궁서 16")     # 위젯 전체 폰트를 궁서 16으로

# 로고 이미지 넣기
lab_logo = Label(win)
img = PhotoImage(file = 'daumLogo.png', master = win)     # PhotoImage()라는 함수를 통해서 이미지 객체 생성
img = img.subsample(2)    # img의 사이즈를 1/2로 줄임
lab_logo.config(image = img)     # image옵션으로 img라는 객체를 집어 넣도록 설정
lab_logo.pack()

# id 레이블
lab1 = Label(win)
lab1.config(text = "ID")
lab1.pack()

# id 입력창
ent1 = Entry(win)
ent1.insert(0, "emailID@browser.com")    # insert()함수로 입력위치와 입력내용을 넣어줌, 0은 시작을 의미
def clear(event):     # event라는 입력변수를 주어서, event가 발생시에만 이 함수 실행되도록 설정
    if ent1.get() == "emailID@browser.com":    # 입력값이 맨 처음 삽입해놓은 내용일 때만, 삭제하도록
        ent1.delete(0, len(ent1.get()) )     #ent1에 delete()함수 실행, 몇번째 글자에서(0) 몇번째 글자까지(입력전체) 지운다.

ent1.bind("<Button-1>", clear)     # bind()함수는 지정한 이벤트(<Button-1>)가 발생되면, 함수(clear)를 실행하도록 함
ent1.pack()

# pw 레이블
lab2 = Label(win)
lab2.config(text = "Password")
lab2.pack()

# pw 입력창
ent2 = Entry(win)
ent2.config(show = '*')    # show옵션은 지정한 문자열(*)로 입력을 변환함
ent2.pack()

# 로그인 버튼
btn = Button(win)
btn.config(text = "로그인")
def login():
    my_id = ent1.get()
    my_pw = ent2.get()
    print(my_id, my_pw)     # 로그인 버튼을 눌렀을 때, 코드창에 프린트되게 설정해봄
    lab3.config(text = "[메시지] 로그인 성공")     # 처음 아무 내용없던 레이블에 함수의 윗부분이 실행되면 레이블 텍스트 보여짐
    
btn.config(command = login)     # 버튼이 클릭되었을 때 command옵션으로 login()이라는 함수를 실행하도록 설정함
btn.pack()

# 로그인 성공 메시지 레이블
lab3 = Label(win)
lab3.pack()




win.mainloop()
```
