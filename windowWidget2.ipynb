```python
from tkinter import*

window = Tk()

## 화면 구성하고 처리하는 부분 ##
photo1 = PhotoImage(file ="gif/cat.gif")
label1 = Label(window, image = photo1)
label1.pack(side=LEFT)

photo2 = PhotoImage(file ="gif/cat2.gif")
label2 = Label(window, image = photo2)
label2.pack(side=LEFT)



window.mainloop()
```
