```python
from tkinter import *
from tkinter import messagebox

root = Tk()    #윈도우 변수 root
root.title('tkinter messagebox')

def popup():
    messagebox.showinfo("팝업창", "Hello World")    # 팝업창 타이틀과 내용

# 그외 showwarning(워닝 사운드도 같이 울림), showerror(에러 사운드), askquestion, askokcancel, askyesno
    
Button(root, text = "팝업", command = popup).pack()

mainloop()

```

```python
from tkinter import *
from tkinter import messagebox

root = Tk()    #윈도우 변수 root
root.title('tkinter messagebox')

def popup():
    response = messagebox.askyesno("팝업창", "Hello World")    # 팝업창 타이틀과 내용
    #Label(root, text = response).pack()  
    if response == 1:
        Label(root, text = "예를 눌렀습니다.").pack()
    else:
        Label(root, text = "아니요를 눌렀습니다." ).pack()

    
Button(root, text = "팝업", command = popup).pack()

mainloop()
```
