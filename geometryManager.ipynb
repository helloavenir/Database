```python
from tkinter import *

win = Tk()
f = Frame(win)      #win이라는 윈도우 창 안에 프레임을 만든다.

b1 = Button(f, text="박스1", bg="red", fg="white")    #f라는 프레임 안에 버튼을 만든다.
b2 = Button(f, text="박스2", bg="orange", fg="gray")
b1.pack(side=LEFT)
b2.pack(side=LEFT)

l = Label(win, text="이 레이블은 버튼들 위 윈도우 창에 배치된다.")    #l이라는 레이블을 윈도우 창에 만든다.
l.pack()    #l이라는 레이블을 기본 윈도우(루트 윈도우)에 압축 배치 관리자로 배치한다.
f.pack()    #f이라는 프레임을 기본 윈도우(루트 윈도우)에 압축 배치 관리자로 배치한다.

win.mainloop()
```
